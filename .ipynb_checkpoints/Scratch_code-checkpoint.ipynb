```python
## lets try out the geographies and sectors...

## load in the ULEZ 

#post codes from TfL: https://tfl.gov.uk/corporate/transparency/freedom-of-information/foi-request-detail?referenceId=FOI-2241-1819

import pandas as pd # version 2.1.4
import os
import seaborn as sns
import glob
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read in to get the postcode sectors in london 

postcode = pd.read_csv('../Data/avgPrice-county-GREATER_LONDON-by-pcSector-any-2023.csv', names = ['postcode'],skiprows = 6, usecols = [0])
```

```python
london_postcodes = postcode['postcode'].tolist()
london_folder = [p.split(' ')[0] for p in london_postcodes]
london_folder = np.unique(london_folder)
```

```python
# code from here: https://gist.github.com/ericrobskyhuntley/790937a759fd89ed77c8831f880f854c

#Create a unified geojson 

import os
import glob
import geojson


#limitation: postcodes are not defined by the government as polygons, therefore third parties have created polygons from the points. 
#Geojsons for each post code downloaded from here: https://postcodes.mapit.longair.net/
json_dir_name = "../Data/gb-postcodes-v5/sectors"
list_of_london_folders = london_folder #os.listdir(json_dir_name) dont need all, just the london ones

list_of_london_folder_paths = [json_dir_name + '/' + f for f in list_of_london_folders]

file_list_london = []
for f in list_of_london_folder_paths:     
    json_pattern = os.path.join(f,'*.geojson')
    file_list = glob.glob(json_pattern)
    file_list_london.extend(file_list)


```

```python
print(len(list_of_london_folder_paths))
```

```python
import geopandas as gpd
```

```python
import geoplot
import geoplot.crs as gcrs
```

```python
list_df = []
for i in file_list_london: 
    df = gpd.read_file(i)
    list_df.append(df)
    
```

```python
london_geo = pd.concat(list_df)
```

```python
london_geo.shape
```

```python
london_geo.crs
```

```python
geoplot.polyplot(london_geo, projection=gcrs.AlbersEqualArea(), edgecolor='darkgrey', facecolor='lightgrey', linewidth=.3,
    figsize=(12, 8))
geoplot.polyplot(ulez_geo, projection=gcrs.AlbersEqualArea(), edgecolor='darkblue', linewidth=.3,
    figsize=(12, 8))
```

```python
#show postcode sectors and the ULEZ compared to it... 

base = london_geo.plot(color='white', edgecolor='black')
ulez_geo.plot(ax=base, marker='o', color='red', markersize=5, alpha= 0.5)
```

```python
#next, get a group of sectors that intersect or border the ulez ... covered_by
london_geo.intersects(ulez_geo)

intersection = london_geo[london_geo.intersects(ulez_geo)]

covered = london_geo[london_geo.covered_by(ulez_geo)]
```
